```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def addNumbers(a:int , b:int):
    """
    This function will return the sum of two numbers
    Args:
        a:Number one
        b:Number two
    """
    return a + b


@tool
def multiplyNumbers(a:int , b:int):
    """
    This function will return the product of two numbers
    Args:
        a:Number one
        b:Number two
    """
    return a * b
```

```python
addNumbers.invoke({"a":12,"b":3})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_agent
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
agent = create_agent(
    model=llm,
    tools=[addNumbers , multiplyNumbers],
    system_prompt= "You are a math teacher and always use tools for calculation"
)
```

```python
response = agent.invoke({"messages":[{"role":"user" , "content":"What is 2 + 3 * 100"}]})
response["messages"][-1].content

```
