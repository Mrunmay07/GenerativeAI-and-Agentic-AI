```python
from pydantic import BaseModel , Field
from typing import List, Annotated
```

```python
class User(BaseModel):
    name : str = ""
    email : str = ""
    age : int = 0
    
```

```python
user = User(name="Mayur", age = 24)
user
```

# Field Validation

```python
class Product(BaseModel):
    title : str = Field(max_length=10)
    description : str = Field(min_length=20)
    price : int  = Field(gt=0, le=100)
    
```

```python
product = Product(title="Samsung", description="This is the flagship mobile from samsung" ,price=100 )

product2 = Product(title="Iphone", description="This is the flagship mobile from apple" ,price=56 )

product
```

# Multiple Class Validation

```python
class Order(BaseModel):
    id : str = Field(min_length=6 , max_length=12)
    product : Product
    
    
```

```python
order = Order(id="nd34512" , product=product)
order.product.title
```

```python
class Carts(BaseModel):
    id: str = Field(min_length=6 , max_length=12)
    items : List[Product]
    comment : str = Field(default="No Comment")
```

```python
cart = Carts(id = "nn34as13d" , items = [product , product2])


for product in cart.items:
    print(product.title)
    
```

# Annotated

```python
commonValidator = Annotated[List[Product] , Field(default = [])]
```

```python
class Carts(BaseModel):
    id: str = Field(min_length=6 , max_length=12)
    items : commonValidator
    comment : str = Field(default="No Comment")
```

```python
class Order(BaseModel):
    products : commonValidator
    
```

```python
ord = Order(products=[product , product2])
ord
```
