```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model="openai/gpt-oss-20b" , streaming=True)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite" , streaming=True)
```

```python
question = "What is Reconcilliaton ? Explain with React fiber in detail , indepth"
```

```python
res = llm.stream(question)
res
```

```python
res = llm_gemini.stream(question)
for chunk in res:
    print(chunk.content,end="")
```
